# Data Preprocessing

```python
# Import libraries
import os 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL import Image
```

### Get the cardinality of the data that we are going to use for the model

```python
# Iterate through the training folder and get the cardinality of the data classes
train_path = "archive/Training"
categories = ["glioma", "meningioma", "notumor", "pituitary"]


# Initialize counters for files in each category
train_counts = {category: 0 for category in categories}

# Traverse through the directory structure
for root, dirs, files in os.walk(train_path):
    for file in files:
        for category in categories:
            if category in root:
                train_counts[category] += 1


# Extract the counts and categories
counts = [train_counts[category] for category in categories]
categories = [category.capitalize() for category in categories]


# Create the bar chart
plt.bar(categories, counts, color=(0.2, 0.4, 0.6, 0.6))
plt.xlabel("Brain Tumor Categories")
plt.ylabel("Number of Files")
plt.title("Number of Files per Category Training Data-set")
plt.show()    
```

```python
# Iterate through the training folder and get the cardinality of the data classes
test_path = "archive/Testing"
test_categories = ["glioma", "meningioma", "notumor", "pituitary"]


# Initialize counters for files in each category
test_counts = {category: 0 for category in test_categories}

# Traverse through the directory structure
for root, dirs, files in os.walk(test_path):
    for file in files:
        for category in test_categories:
            if category in root:
                test_counts[category] += 1


# Extract the counts and categories
counts = [test_counts[category] for category in test_categories]
categories = [category.capitalize() for category in test_categories]


# Create the bar chart
plt.bar(categories, counts)
plt.xlabel("Brain Tumor Categories")
plt.ylabel("Number of Files")
plt.title("Number of Files per Category Testing Data-set")
plt.show()    
```

### Check the sizes of the images for each category

```python
# Define the path to the Archive folder
training_path = "archive/Training/"

# Initialize lists to store image sizes
heights = []
widths = []

# Traverse through the folder structure
for root, _, files in os.walk(training_path):
    # Get the folder name
    folder_name = os.path.basename(root)
    
    # Check if it's one of the four target folders
    if folder_name in ["glioma", "meningioma", "notumor", "pituitary"]:

        # Iterate over the files in the folder
        for file in files:            
            if file.endswith(".jpg") or file.endswith(".png"):
                
                # Get the image path, open the image with PIL and get the size
                image_path = os.path.join(root, file)                
                image = Image.open(image_path)
                width, height = image.size
                
                # Append the image sizes to the lists
                heights.append(height)
                widths.append(width)



# Create the scatter plot
plt.scatter(widths, heights, color=(0.2, 0.4, 0.6, 0.6))
plt.grid()
plt.xlabel("Width")
plt.ylabel("Height")
plt.title("Training Image Sizes")
plt.show()
```
